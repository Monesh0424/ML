```python
from sklearn.base import BaseEstimator , TransformerMixin
```

```python
class RemovingSpaces(BaseEstimator,TransformerMixin):
    def __init__(self,columns=True):
        self.columns = columns
        
    def fit(self,X,y=None):
        return self
    
    def transform(self,X,y=None):
        col = list(X.columns)
        li = []
        for i in col:
            li.append(i.strip())
        return pd.DataFrame(data=X.values,columns=li,index=X.index)
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics
import joblib
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor 
from sklearn.svm import SVR 
```

```python
def run_model(feat , labels , name):
    
        # Building linear regression model
        if name == 'lr':
            lm = LinearRegression()
            lm.fit(feat,labels)
            pred_lm = lm.predict(feat)
            rmse_lm = np.sqrt(metrics.mean_squared_error(labels,pred_lm))
            print("rmse for lr : {}".format(rmse_lm))
            print('\n')
            print("r2 score :",metrics.r2_score(labels,pred_lm))
            print('\n')
        # Evaluating lm model
            lm_scores = cross_val_score(estimator = lm , X = feat , y = labels , scoring = 'neg_mean_squared_error' , cv=10)
            lm_rmse_scores = np.sqrt(-lm_scores)
            print('scores :' ,list(lm_scores))
            print('\n')
            print('Mean :' , lm_scores.mean())
            print('\n')
            print('std :' , lm_scores.std())
        
        #-------------------------------------------------------------------------------------------
        
        # Building Decision tree regressor
        
        if name == 'dtr':
            
            dtree = DecisionTreeRegressor()
            dtree.fit(feat,labels)
            pred_tree = dtree.predict(feat)
            rmse_tree = np.sqrt(metrics.mean_squared_error(labels,pred_tree))
            print("rmse for dtr : {}".format(rmse_tree))
            print('\n')
        # Evaluating tree model
            tree_scores = cross_val_score(estimator = dtree , X = feat , y = labels , scoring = 'neg_mean_squared_error' , cv=10)
            tree_rmse_scores = np.sqrt(-tree_scores)
            print('scores :' ,tree_scores)
            print('\n')
            print('Mean :' , tree_scores.mean())
            print('\n')
            print('std :' , tree_scores.std())

        #-------------------------------------------------------------------------------------------
        
        # Building random forest regressor
        if name == 'rfr':
            forest = RandomForestRegressor()
            forest.fit(feat,labels)
            pred_forest = forest.predict(feat)
            rmse_forest = np.sqrt(metrics.mean_squared_error(labels,pred_forest))
            print("rmse for dtr : {}".format(rmse_forest))
            print('\n')
        
        # Evaluating forest model
            forest_scores = cross_val_score(estimator = forest , X = feat , y = labels , scoring = 'neg_mean_squared_error' , cv=10)
            forest_rmse_scores = np.sqrt(-forest_scores)
            print('scores :' ,list(forest_scores))
            print('\n')
            print('Mean :' , forest_scores.mean())
            print('\n')
            print('std :' , forest_scores.std())

        #-------------------------------------------------------------------------------------------
        
        # Building support vector regressor model
        if name == 'svr':
            svr = SVR()
            svr.fit(feat,labels)
            pred_svr = svr.predict(feat)
            rmse_svr = np.sqrt(metrics.mean_squared_error(labels,pred_svr))
            print("rmse for svr : {}".format(rmse_svr))
            print('\n')
            
        # Evaluating svr model
            svr_scores = cross_val_score(estimator = svr , X = feat , y = labels , scoring = 'neg_mean_squared_error' , cv=10)
            svr_rmse_scores = np.sqrt(-svr_scores)
            print('scores :' ,list(svr_scores))
            print('\n')
            print('Mean :' , svr_scores.mean())
            print('\n')
            print('std :' , svr_scores.std())
```

```python
decision_tree_reg  = run_model(final_training_data , train_data_labels , 'dtr')
```
